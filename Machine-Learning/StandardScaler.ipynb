```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Social_Network_Ads.csv")
df = df.iloc[:,2:]
df.head()
```

```python
df.describe()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('Purchased',axis = 1), df['Purchased'], test_size = 0.3, random_state=0)

X_train.shape, X_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

# fit the scaler to train set
scaler.fit(X_train)

#transform train and test sets
X_scaled_train = scaler.transform(X_train)
X_scaled_test = scaler.transform(X_test)
```

```python
scaler.mean_
```

```python
X_scaled_train = pd.DataFrame(X_scaled_train, columns=X_train.columns)
X_scaled_test = pd.DataFrame(X_scaled_test, columns=X_test.columns)
```

```python
np.round(X_train.describe(),1)
```

```python
np.round(X_scaled_train.describe(),1)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr_scaled = LogisticRegression()
```

```python
y_train = y_train.astype("int")
```

```python
lr.fit(X_train, y_train)
lr_scaled.fit(X_scaled_train, y_train)
```

```python
y_pred = lr.predict(X_test)
y_scaled_pred = lr_scaled.predict(X_scaled_test)
```

```python
from sklearn.preprocessing import LabelEncoder

# Convert y_pred to multiclass
le = LabelEncoder()
le.fit(y_test)
y_pred = le.transform(y_pred)

# Convert y_scaled_pred to multiclass
le.fit(y_test)
y_scaled_pred = le.transform(y_scaled_pred)
```

```python
from sklearn.metrics import accuracy_score
print('Actual score: ', accuracy_score(y_test, y_pred))
print('Scaled score: ', accuracy_score(y_test, y_scaled_pred))
```
